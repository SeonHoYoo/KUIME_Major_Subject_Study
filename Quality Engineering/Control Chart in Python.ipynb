```python
data = np.array([2.49, 3.39, 7.41, 2.88, 0.76, 1.32, 7.05, 1.37, 6.17, 5.12,
                 1.34, 0.50, 4.35, 1.67, 1.63, 4.88, 15.19, 0.67, 4.14, 2.16,
                 1.14, 2.66, 4.67, 1.54, 5.06, 3.40, 1.39, 2.67, 6.92, 36.99])
```

```python
mean_x = np.mean(data)
sigma_x = np.std(data, ddof=1)

mr = np.abs(np.diff(data))  # 이동 범위 계산
mean_mr = np.mean(mr)

# 이동 범위 관리도 상수 (표본 크기가 2일 때)
D4 = 3.267
D3 = 0
D2 = 1.128

UCL_x = mean_x + 3 * (mean_mr)/D2
LCL_x = mean_x - 3 * (mean_mr)/D2

UCL_mr = D4 * mean_mr
LCL_mr = D3 * mean_mr  # LCL은 음수가 될 수 없으므로 0
```

```python
print(f"mean: {mean_x}")
print(f"sigma_x: {sigma_x}")
print(f"UCL_x: {UCL_x}")
print(f"LCL_x: {LCL_x}")
print(f"mean_mr: {mean_mr}")
print(f"UCL_mr: {UCL_mr}")
print(f"LCL_mr: {LCL_mr}")
```

```python
# x chart 그리기
plt.figure(figsize=(12, 6))
plt.plot(data, marker='o', linestyle='-')
plt.axhline(UCL_x, color='r', linestyle='--', label='UCL (3σ)')
plt.axhline(mean_x, color='g', linestyle='-', label='Center Line (Mean)')
plt.axhline(LCL_x, color='r', linestyle='--', label='LCL (3σ)')
plt.title('x control Chart')
plt.xlabel('Observation')
plt.ylabel('Waiting Time')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(mr, marker='o', linestyle='-')
plt.axhline(UCL_mr, color='r', linestyle='--', label='UCL (MR)')
plt.axhline(mean_mr, color='g', linestyle='-', label='Center Line (Mean MR)')
plt.axhline(LCL_mr, color='r', linestyle='--', label='LCL (MR)')
plt.title('Moving Range (MR) Control Chart')
plt.xlabel('Observation')
plt.ylabel('Moving Range')
plt.show()
```

### 7.3

```python
sample_size = 50
nonconforming_units = np.array([0, 3, 4, 6, 5, 2, 8, 9, 4, 2, 6, 4, 8, 0, 7, 20, 6, 1, 5, 7])
```

```python
# 불량 비율 계산
p = nonconforming_units / sample_size

# 평균 불량률 계산
p_bar = np.mean(p)

# 관리 한계 계산
UCL = p_bar + 3 * np.sqrt((p_bar * (1 - p_bar)) / sample_size)
LCL = p_bar - 3 * np.sqrt((p_bar * (1 - p_bar)) / sample_size)
```

```python
print(f"p_bar: {p_bar}")
print(f"UCL: {UCL}")
print(f"LCL: {LCL}")
```

```python
# p-chart 그리기
plt.figure(figsize=(10, 6))
plt.plot(p, marker='o', linestyle='-', color='b', label='Sample p')
plt.axhline(UCL, color='r', linestyle='--', label='UCL')
plt.axhline(p_bar, color='g', linestyle='-', label='Center Line (p_bar)')
plt.axhline(LCL, color='r', linestyle='--', label='LCL')
plt.title('p-Chart for Nonconforming Units')
plt.xlabel('Day')
plt.ylabel('Fraction Nonconforming (p)')
plt.show()
```

### 7.23

```python
sample_size = 100
nonconforming_units = np.array([10, 15, 31, 18, 24, 12, 23, 15, 8, 8])
```

```python
# 불량 비율 계산
p = nonconforming_units / sample_size

# 평균 불량률 계산
p_bar = np.mean(p)

# 관리 한계 계산
UCL = p_bar + 3 * np.sqrt((p_bar * (1 - p_bar)) / sample_size)
LCL = p_bar - 3 * np.sqrt((p_bar * (1 - p_bar)) / sample_size)

# LCL이 0보다 작으면 0으로 설정
LCL = max(LCL, 0)
```

```python
print(f"p_bar: {p_bar}")
print(f"UCL: {UCL}")
print(f"LCL: {LCL}")
```

```python
# p-chart * number of sample = number nonconforming chart
plt.figure(figsize=(10, 6))
plt.plot(sample_size * p, marker='o', linestyle='-', color='b', label='Sample p')
plt.axhline(sample_size * UCL, color='r', linestyle='--', label='UCL')
plt.axhline(sample_size * p_bar, color='g', linestyle='-', label='Center Line (p_bar)')
plt.axhline(sample_size * LCL, color='r', linestyle='--', label='LCL')
plt.title('Number Nonconforming Chart')
plt.xlabel('Sample Number')
plt.ylabel('Fraction Nonconforming (p)')
plt.show()
```
